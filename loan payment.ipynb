```python
# EXPORT THE TRAIN DATASET
train_data = pd.read_csv("application_train.csv")
train_data.head()
```

```python
# DESCRIBING THE TRAIN DATASET
train_data.describe()
```

```python
#CHECKING FOR THE MISSING DATA
train_data.isna().sum()
```

```python
# DROPPING THE MISSING DATASET
train_data = train_data.dropna()
```

```python
# RECHECKING IF THERE IS ANY MISSING DATASET
train_data.isna().sum()
```

```python
# ASSIGNING MY X AND Y VARIABLE USING THE TRAIN DATASET
x = train_data.drop("TARGET", axis = 1)
y = train_data["TARGET"]
```

```python
# IMPORTING FROM SKLEARN MODEL
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# CONVERTION FROM NON-NUMERICAL DATA TO NUMERICAL DATA
text_data = ["NAME_CONTRACT_TYPE","CODE_GENDER","FLAG_OWN_CAR","FLAG_OWN_REALTY","NAME_TYPE_SUITE",
             "NAME_INCOME_TYPE","NAME_EDUCATION_TYPE","NAME_FAMILY_STATUS","NAME_HOUSING_TYPE","FONDKAPREMONT_MODE",
             "OCCUPATION_TYPE","HOUSETYPE_MODE","WEEKDAY_APPR_PROCESS_START","ORGANIZATION_TYPE","WALLSMATERIAL_MODE","EMERGENCYSTATE_MODE"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, text_data)], remainder = "passthrough")
transformed_x = transformer.fit_transform(x)
```

```python
# THE OUTPUT OF TRANSFORMED DATA
transform_x = transformer.fit_transform(x)
transform_x
```

```python
# FITTING THE NUMERIC DATA BACK 
dummie = pd.get_dummies(train_data[["NAME_CONTRACT_TYPE","CODE_GENDER","FLAG_OWN_CAR","FLAG_OWN_REALTY","NAME_TYPE_SUITE",
             "NAME_INCOME_TYPE","NAME_EDUCATION_TYPE","NAME_FAMILY_STATUS","FONDKAPREMONT_MODE","HOUSETYPE_MODE",
             "OCCUPATION_TYPE","WEEKDAY_APPR_PROCESS_START","ORGANIZATION_TYPE","WALLSMATERIAL_MODE","EMERGENCYSTATE_MODE"]])
dummie.head()
```

```python
# USING OF CLASSIFICATION MODEL OF SKLEARN
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
# TRAINING MY DATA
np.random.seed(42)
x_test,x_train,y_test,y_train = train_test_split(transform_x,y, test_size = 0.2)
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_preds = model.predict(x_test)
```

```python
# PREDICTING MY MODEL
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_preds)
```
